```python
import pandas as pd
import numpy as np
import nltk
from nltk import word_tokenize
```

```python
df = pd.read_csv(r"D:\Datasets\bbc_text_cls.csv")
```

```python
idx = 0
word2idx = {}
tokenized_docs = []
for doc in df['text']:
    words = word_tokenize(doc.lower())
    doc_as_int = []
    for word in words:
        if word not in word2idx:
            word2idx[word] = idx
            idx += 1
        
        doc_as_int.append(word2idx[word])
    tokenized_docs.append(doc_as_int)

        
```

```python
print(len(word2idx))
print(word2idx)
```

```python
print("How tokenized docs look :",tokenized_docs[:11])
print("Length of tokenized docs :",len(tokenized_docs))
print("Length of first element in tokenized docs :",len(tokenized_docs[0]))
```

```python
idx2word = {v:k for k,v in word2idx.items()}
print(len(idx2word))
print(idx2word)
```

```python
n = len(df['text'])
v = len(word2idx)
```

```python
# instantiating term-frequency matrix

tf = np.zeros((n, v))
```

```python
for i, doc_as_int in enumerate(tokenized_docs):
  for j in doc_as_int:
    tf[i, j] += 1
```

```python
document_frequency = np.sum(tf > 0 , axis=0)
idf = np.log(n/document_frequency)
```

```python
tfidf = tf * idf
```

```python
np.random.seed(123)
```

```python
i = np.random.choice(n)
row = df.iloc[i]
print( "Label :", row['labels'])
print("Text :", row['text'].split("\n",1)[0])
print("Top 5 terms :")

scores = tfidf[i]
indices = (-scores).argsort()

for j in indices[:5]:
    print(idx2word[j])

```
